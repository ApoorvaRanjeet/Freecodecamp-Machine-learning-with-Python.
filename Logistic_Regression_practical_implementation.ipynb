```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species'].unique()
```

```python
# to check the null value in the datset
df.isnull().sum()
```

```python
# to remove any category from the output
df['species']!='setosa'
```

```python
df = df[df['species']!='setosa'] # jitne bhi column hai setosa ke woh saare remove ho jayenge
```

```python
df.head()
```

```python
# to represent the putputs in numeric format i.e. 0 and 1
df['species'] = df['species'].map({'versicolor':0,'virginica':1})
```

```python
df.head()
```

```python
# split the dataset into independent and dependent features
# x=df.iloc[:,:-1]
# y=df.iloc[:,-1]
# another method to split the dataset
```

```python
X = df.drop(columns='species',axis=1)
Y = df['species']
```

```python
from sklearn.model_selection import train_test_split
(x_train,x_test,y_train,y_test) = train_test_split(X,Y,test_size=.30,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'penalty':['l2'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
 # classification problem ka scoring hamesha accuracy hoga , regerssion ka hamesha neg_mean_squared_error hoga
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
# prediction

y_pred = classifier_regressor.predict(x_test)
```

```python
# accuracy score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
score
```
