Importing the dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

Loading the dataset

```python
dataset = pd.read_csv('/content/data.csv') # ,index_column=0 this will automatically remove the null column from the dataset
```

```python
# to check the missing values in each column
dataset.isnull().sum()
```

```python
# checking the districution of the traget variable
dataset['diagnosis'].value_counts()
```

```python
dataset= pd.get_dummies(dataset,columns=['diagnosis'],drop_first=True)
dataset= pd.get_dummies(dataset,columns=['Unnamed: 32'],drop_first=True)
```

```python
dataset
```

```python
# Separating the features and taregt
X = dataset.drop(columns='diagnosis_M',axis=1)
Y = dataset['diagnosis_M']
```

```python
X
```

```python
# splittint the train and test data
(x_train,x_test,y_train,y_test) = train_test_split(X,Y,test_size=.20,random_state=2)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
y_test
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

Building the basic Neural network

Importing the libraries

```python
import tensorflow as tf
# tf.random.set_seed(3)
from tensorflow import keras
```

Build the model

```python
# setting up the layers of neural network
model = keras.Sequential(
    [
        keras.layers.Flatten(input_shape=(31,)),
        keras.layers.Dense(20,activation='relu'),
        keras.layers.Dense(2,activation='sigmoid')
    ]
)
```

```python
model.summary()
```

Compile the model

```python
model.compile(optimizer = 'adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

Training the model

```python
model.fit(x_train,y_train,epochs=50)
```

```python
# evaluate the model
loss,accuracy = model.evaluate(x_test,y_test)
```

```python
print(accuracy)
```

```python
prediction = model.predict(x_test)
```
