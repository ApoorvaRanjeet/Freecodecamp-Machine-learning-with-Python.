```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.utils import load_img, img_to_array
```

```python
img=load_img('/content/cat-551554_640.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img)
```

PIL : this is python image library object

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

we have created the object of the class ImageDataGenerator which we have imported earlier.


```python
img=img_to_array(img)
```

here we converted the type of image that was PIL into the numpy array by using the function img_to_array

```python
type(img)
```

```python
input_batch = img.reshape(1,200,200,3)
```

The reshape operation is creating a new array with the same data but in a new shape. The first dimension is set to 1, which usually indicates that the array is now representing a batch of images, where each image has a size of 200x200 pixels and three color channels.

```python
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='/content/aug'):
  i=i+1
  if i==10:
    break
```

we are using calling this object datagen which has this function called flow which is used when transformation is being performed on single image
