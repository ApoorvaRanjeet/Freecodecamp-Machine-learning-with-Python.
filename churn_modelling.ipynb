```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
# creating the dummy variables
dataset=pd.get_dummies(dataset,columns=['Geography','Gender'],drop_first=True) # one-hot encoding
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['Exited'].value_counts()
```

```python
dataset['Surname'].value_counts()
```

```python
# drop unnecessary columns
dataset = dataset.drop(['Geography_Germany','Geography_Spain','Gender_Male','Surname'],axis=1)
dataset
```

```python
# separating the target and features
X = dataset.drop(columns='Exited',axis=1)
Y = dataset['Exited']
```

```python
X
```

```python
# splitting the train test data
(x_train,x_test,y_train,y_test) = train_test_split(X,Y,test_size=.30,random_state=40)
```

```python
x_test
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#building the model
import tensorflow as tf
from tensorflow import keras
```

```python
model =  keras.Sequential([
    keras.layers.Flatten(input_shape=(10,)),
    keras.layers.Dense(6,activation='relu'),
    keras.layers.Dense(4,activation='sigmoid'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# training the model
model.fit(x_train,y_train,validation_split=.20,batch_size=5,epochs=100)
```

```python
#evaluate
loss,accuracy = model.evaluate(x_test,y_test)
```

```python
accuracy
```

```python
#predicting
prediction = model.predict(x_test)
```
